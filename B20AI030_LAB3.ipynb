```python
# load the data into a GeoDataFrame
flight_data = pd.read_csv("/content/Flightschedule .csv")

flight_data
```

```python
flight_data.dtypes
```

```python
coordinates = pd.read_csv('/content/airport_cordinates.csv')
coordinates
```

```python
# modifying the data to get only unique flights details
unique_flights = flight_data[['Origin', 'Destination', 'Flight Number']].drop_duplicates()

# merging the coordinates of the unique flights of both the origin airport and destination airport
unique_flights = pd.merge(unique_flights, coordinates[['city', 'latitude', 'longitude']], left_on='Origin', right_on='city', how='left')
unique_flights = unique_flights.rename(columns={'latitude': 'origin_latitude', 'longitude': 'origin_longitude'}).drop(columns=['city'])

unique_flights = pd.merge(unique_flights, coordinates[['city', 'latitude', 'longitude']], left_on='Destination', right_on='city', how='left')
unique_flights = unique_flights.rename(columns={'latitude': 'destination_latitude', 'longitude': 'destination_longitude'}).drop(columns=['city'])

print(unique_flights.head())
```

```python
unique_flights['geometry'] = [LineString([(row['origin_longitude'], row['origin_latitude']),
                                          (row['destination_longitude'], row['destination_latitude'])])
                              for _, row in unique_flights.iterrows()]

# creating a GeoDataFrame for all flight connections
unique_flights_df = gpd.GeoDataFrame(unique_flights, geometry='geometry')

# creating a GeoDataFrame for all the airports
airports_df = gpd.GeoDataFrame(coordinates, geometry=gpd.points_from_xy(coordinates['longitude'], coordinates['latitude']))

# plotting the connections and airports on a map
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())

# add countries as a basemap
ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle='-', alpha=.5)
ax.add_feature(cfeature.STATES.with_scale('10m'))

# plot the connections
unique_flights_df.plot(ax=ax, color='blue', linewidth=1)

# plot the airports and annotate them with their index
airports_df.plot(ax=ax, color='red', markersize=5, marker='*', label='airports')

for i, row in airports_df.iterrows():
    ax.annotate(text=str(i), xy=(row.geometry.x, row.geometry.y),
                xytext=(5, 5), textcoords='offset points', fontsize=8)

# set the title of the plot
ax.set_title('Flight Routes')
plt.show()
```
