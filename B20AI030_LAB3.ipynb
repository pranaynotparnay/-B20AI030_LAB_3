```python
df=pd.read_csv('population.csv')
```

```python
#Merging the data
india = india.set_index('NAME_2').join(df.set_index('District_of_Capital'))
india['Population'] = india['Population'].replace(np.nan, 0)
```

```python
india['NAME_2'] = india.index
```

```python
india=india.reset_index()
```

```python
india.loc[india["NAME_2"] == "Patna", "Population"] = 2481530
india.loc[india["NAME_2"] == "Delhi", "Population"] = 15217000
india.loc[india["NAME_2"] == "Hyderabad", "Population"] = 10268653
india.loc[india["NAME_2"] == "Lucknow", "Population"] = 3764619
india.loc[india["NAME_2"] == "Delhi", "Capital"] = 'Delhi'
india.loc[india["NAME_2"] == "Delhi", "State_Name"] = 'Delhi'
```

```python
#india=india[india.Population != 0]
```

```python
india.rename(columns = {'index':'District_of_Capital'}, inplace = True)
```

```python
india = india.sort_values("State_Name")
```

```python
india
```

```python
india.dtypes
```

```python
import plotly_express as px
```

```python
# Create a figure and axis
fig, ax = plt.subplots(figsize=(16, 16),
                       subplot_kw={'projection': ccrs.PlateCarree()})

# Plot the map of India
india.plot(ax=ax, color='lightgrey', edgecolor='black', linewidth=0.4)

# Plot the bubble map of state capital populations
india.plot(ax=ax, column='Population', 
           cmap='YlOrRd', markersize=[39.739192, -104.990337],             
           alpha=0.7, legend=True,legend_kwds={'label': "Populations"})

# Add title and labels
ax.set_title('State Capital Populations in India', fontsize=20)

# Show the plot
plt.show()
```

# **Interactive Bubble Map**

```python
centroids = india.copy()
centroids.geometry = india.centroid
centroids['size'] = centroids['Population'] / 5000  # to get reasonable plotable number

#subplots=plt.subplots(subplot_kw={'projection': ccrs.PlateCarree()}, legend=True,legend_kwds={'label': "Populations"})

ax = india.plot(figsize=(16,16),facecolor='w', edgecolor='k',alpha=0.7, linewidth=0.4)
centroids.plot(markersize='size', ax=ax)
# Add title and labels
ax.set_title('State Capital Populations in India', fontsize=20)
```
